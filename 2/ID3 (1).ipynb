```python
import numpy as np

class Node:
    def __init__(self, attribute):
        self.attribute = attribute
        self.children = {}

    def add_child(self, value, child_node):
        self.children[value] = child_node

def entropy(data):
    classes = np.unique(data[:, -1])
    entropy_value = 0
    total_samples = len(data)

    for c in classes:
        class_instances = len(data[data[:, -1] == c])
        p = class_instances / total_samples
        entropy_value -= p * np.log2(p)

    return entropy_value

def information_gain(data, attribute, threshold=None):
    if threshold is not None:
        lower_split = data[data[:, attribute] <= threshold]
        upper_split = data[data[:, attribute] > threshold]
        splits = [lower_split, upper_split]
    else:
        splits = np.split(data, np.unique(data[:, attribute], return_index=True)[1])[1:]

    total_samples = len(data)
    new_entropy = sum(len(split) / total_samples * entropy(split) for split in splits)
    return entropy(data) - new_entropy

def find_best_split(data, attributes):
    best_gain = 0
    best_attribute = None
    best_threshold = None

    for attribute in attributes:
        unique_values = np.unique(data[:, attribute])
        for value in unique_values:
            gain = information_gain(data, attribute, threshold=value)
            if gain > best_gain:
                best_gain = gain
                best_attribute = attribute
                best_threshold = value

    return best_attribute, best_threshold

def build_decision_tree(data, attributes):
    # If all samples have the same class label, return a leaf node with that label
    if len(np.unique(data[:, -1])) == 1:
        return Node(data[0, -1])

    # If there are no more attributes to split on, return a leaf node with the majority class label
    if len(attributes) == 0:
        majority_class = np.argmax(np.bincount(data[:, -1].astype(int)))
        return Node(majority_class)

    # Find the best attribute and threshold (for numerical attributes) to split on
    best_attribute, best_threshold = find_best_split(data, attributes)

    # Create a decision node with the best attribute
    decision_node = Node(best_attribute)

    if best_threshold is not None:
        # For numerical attribute, split the data into two parts based on the threshold
        lower_split = data[data[:, best_attribute] <= best_threshold]
        upper_split = data[data[:, best_attribute] > best_threshold]
        attributes.remove(best_attribute)  # Remove the attribute from future consideration
        decision_node.add_child("<= {}".format(best_threshold), build_decision_tree(lower_split, attributes))
        decision_node.add_child("> {}".format(best_threshold), build_decision_tree(upper_split, attributes))
    else:
        # For categorical attribute, split the data based on each unique value
        unique_values = np.unique(data[:, best_attribute])
        for value in unique_values:
            subset = data[data[:, best_attribute] == value]
            attributes.remove(best_attribute)  # Remove the attribute from future consideration
            decision_node.add_child(value, build_decision_tree(subset, attributes.copy()))

    return decision_node

# Example usage:
# Assuming 'data' is your dataset where the last column is the class label
# Assuming 'attributes' is a list of attribute indices (excluding the class label)
# root_node = build_decision_tree(data, attributes)
```

```python
import pandas as pd
import numpy as np
import math
from graphviz import Digraph

class Node:
    numerical = ['year', 'price', 'mileage', 'tax', 'mpg', 'engineSize']

    def __init__(self, data, edge):
        self.data = data
        self.children = []
        self.is_leaf = False
        self.label = None
        self.edge = edge
        self.target_column_index = self.get_target_column_index()
        self.id3()

    @staticmethod
    def entropy(data):
        classes, counts = np.unique(data, return_counts=True)
        probabilities = counts / len(data)
        entropy_value = -np.sum(probabilities * np.log2(probabilities))
        return entropy_value

    def calculate_entropy(self, data):
        total_rows = len(data)
        target_values = data.unique()
        entropy = 0
        for value in target_values:
            value_count = len(data[data == value])
            proportion = value_count / total_rows
            entropy -= proportion * math.log2(proportion) if proportion != 0 else 0
        return entropy

    def calculate_information_gain_numerical(self, data, feature):
        max_ig = 0
        optimal_threshold = 0
        sorted_data = data.sort_values()
        unique_values = sorted_data.unique()
        for i in range(len(unique_values) - 1):
            threshold = (unique_values[i] + unique_values[i + 1]) / 2
            lower_split = data[data <= threshold]
            upper_split = data[data > threshold]
            ig = self.calculate_entropy(data) - (len(lower_split) / len(data) * self.calculate_entropy(lower_split) + len(upper_split) / len(data) * self.calculate_entropy(upper_split))
            if ig > max_ig:
                max_ig = ig
                optimal_threshold = threshold
        return max_ig, optimal_threshold

    def calculate_information_gain_categorical(self, data, feature):
        unique_values = data.unique()
        weighted_entropy = 0
        for value in unique_values:
            subset = data[data == value]
            proportion = len(subset) / len(data)
            weighted_entropy += proportion * self.calculate_entropy(subset)
        information_gain = self.calculate_entropy(data) - weighted_entropy
        return information_gain, None

    def id3(self):
        if len(self.data.unique()) == 1:
            self.is_leaf = True
            self.label = self.data.iloc[0]
            return

        if len(self.data) == 0:
            self.is_leaf = True
            self.label = self.data.mode().iloc[0]
            return

        max_ig = 0
        best_feature = None
        optimal_threshold = None
        for feature in self.data.columns:
            ig = 0
            if feature in Node.numerical:
                ig, threshold = self.calculate_information_gain_numerical(self.data[feature], feature)
                if ig > max_ig:
                    max_ig = ig
                    best_feature = feature
                    optimal_threshold = threshold
            else:
                ig, _ = self.calculate_information_gain_categorical(self.data[feature], feature)
                if ig > max_ig:
                    max_ig = ig
                    best_feature = feature
        self.label = best_feature
        if best_feature in Node.numerical:
            lower_split = self.data[self.data[best_feature] <= optimal_threshold]
            upper_split = self.data[self.data[best_feature] > optimal_threshold]
            self.children.append(Node(lower_split, f'<= {optimal_threshold}'))
            self.children.append(Node(upper_split, f'> {optimal_threshold}'))
        else:
            unique_values = self.data[best_feature].unique()
            for value in unique_values:
                subset = self.data[self.data[best_feature] == value]
                self.children.append(Node(subset, value))

    def get_target_column_index(self):
        for i, column in enumerate(self.data.columns):
            if column == 'Manufacturer':
                return i
        return 0  # Default to the first column if 'Manufacturer' column is not found

    def visualize(self):
        dot = Digraph()
        def add_nodes(node):
            if node is None:
                return
            if len(node.children) == 0:
                dot.node(str(id(node)), label=str(node.label), fontcolor="red")
            else:
                dot.node(str(id(node)), label=str(node.label))
            for child in node.children:
                add_nodes(child)
                dot.edge(str(id(node)), str(id(child)), label=str(child.edge))
        add_nodes(self)
        try:
            dot.render('tree', format='png', cleanup=True)
            print("Tree visualization saved as tree.png")
        except Exception as e:
            print("Error occurred while rendering the tree:", e)

# Load your dataset into a DataFrame named data
# For example:
data = pd.read_csv('/content/test_ds.csv')

# After constructing the decision tree using the id3 method
root_node = Node(data, None)

# Visualize the decision tree
root_node.visualize()
```

```python
import math
import pandas as pd
from graphviz import Digraph

class Node:
    numerical = ['year', 'price', 'mileage', 'tax', 'mpg', 'engineSize']

    def __init__(self, data, edge):
        self.data = data
        self.children = []
        self.is_leaf = False
        self.label = None
        self.edge = edge
        self.id3()

    def calculate_entropy(self, data):
        target_values = data.iloc[:, 0].value_counts(normalize=True)
        entropy = -sum(p * math.log2(p) for p in target_values)
        return entropy

    def calculate_information_gain_numerical(self, data, feature):
        max_ig = 0
        optimal_threshold = 0
        sorted_data = data.sort_values(by=feature)
        targets = sorted_data.iloc[:, 0]

        for i in range(1, len(sorted_data)):
            print("Index:", i)
            if sorted_data.iloc[i - 1, 0] != sorted_data.iloc[i, 0]:
                threshold = (sorted_data.iloc[i - 1][feature] + sorted_data.iloc[i][feature]) / 2
                ig = self.calculate_information_gain_categorical(data, feature, threshold)

                if ig > max_ig:
                    max_ig = ig
                    optimal_threshold = threshold

        return max_ig, optimal_threshold


    def calculate_information_gain_categorical(self, data, feature, threshold=None):
        if threshold is not None:
            subset_left = data[data[feature] <= threshold]
            subset_right = data[data[feature] > threshold]
        else:
            subset_left = data[data[feature] == data.iloc[0][feature]]
            subset_right = data[data[feature] != data.iloc[0][feature]]

        weight_left = len(subset_left) / len(data)
        weight_right = len(subset_right) / len(data)

        ig = self.calculate_entropy(data) - (weight_left * self.calculate_entropy(subset_left) +
                                             weight_right * self.calculate_entropy(subset_right))

        return ig

    def id3(self):
        if len(self.data.iloc[:, 0].unique()) == 1 or len(self.data.columns) == 1:
            self.is_leaf = True
            self.label = self.data.iloc[:, 0].mode().iloc[0]
            return

        max_ig = 0
        best_feature = None
        optimal_threshold = None

        for feature in self.data.columns[1:]:
            if feature in Node.numerical:
                ig, threshold = self.calculate_information_gain_numerical(self.data, feature)
            else:
                ig = self.calculate_information_gain_categorical(self.data, feature)
                threshold = None

            if ig >= max_ig:
                best_feature = feature
                max_ig = ig
                optimal_threshold = threshold

        self.label = best_feature

        if best_feature in Node.numerical:
            subset_left = self.data[self.data[best_feature] <= optimal_threshold].drop(columns=[best_feature])
            subset_right = self.data[self.data[best_feature] > optimal_threshold].drop(columns=[best_feature])
            self.children.append(Node(subset_left, f'<= {optimal_threshold}'))
            self.children.append(Node(subset_right, f'> {optimal_threshold}'))
        else:
            unique_values = self.data[best_feature].unique()
            for value in unique_values:
                subset = self.data[self.data[best_feature] == value].drop(columns=[best_feature])
                self.children.append(Node(subset, value))

    def visualize(self):
        dot = Digraph()

        def add_nodes(node):
            if node is None:
                return
            if len(node.children) == 0:
                dot.node(str(id(node)), label=str(node.label), fontcolor="red")
            else:
                dot.node(str(id(node)), label=str(node.label))
            for child in node.children:
                add_nodes(child)
                dot.edge(str(id(node)), str(id(child)), label=str(child.edge))

        add_nodes(self)
        try:
            dot.render('tree', format='png', cleanup=True)
            print("Tree visualization saved as tree.png")
        except Exception as e:
            print("Error occurred while rendering the tree:", e)


# Load dataset
data = pd.read_csv('/content/test_ds.csv')

root = Node(data, 0)
root.visualize()
```
