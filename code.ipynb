```python
import os


# Iterate through each image
for image_file in image_files:
    # Split the filename and extension
    filename, extension = os.path.splitext(image_file)
    
    # Check if the extension is not '.jpg'
    if extension != '.jpg':
        # Construct the new filename
        new_filename = filename + '.jpg'
        
        # Ensure the new filename is unique
        counter = 1
        while os.path.exists(os.path.join(dataset_dir, new_filename)):
            new_filename = f"{filename}_{counter}.jpg"
            counter += 1
        
        # Construct the full paths to the old and new files
        old_path = os.path.join(dataset_dir, image_file)
        new_path = os.path.join(dataset_dir, new_filename)
        
        try:
            # Rename the file
            os.rename(old_path, new_path)
            print(f"Renamed {image_file} to {new_filename}")
        except FileNotFoundError:
            print(f"Skipping {image_file}: File not found.")
```
